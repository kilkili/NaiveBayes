```python
dms = pd.get_dummies(df, columns=['workclass','education','sex','marital-status','occupation','native-country'])
```

```python
dms
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
scalar.fit_transform(dms[dms.columns])
```

```python
np.log(dms[dms.columns])
```

```python
x = dms.drop('salary', axis=1)
y = dms['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
naive_bayes = GaussianNB()
naive_bayes.fit(x_train , y_train)
```

```python
train_preds = naive_bayes.predict(x_train)
test_preds = naive_bayes.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
confusion_matrix(y_train, train_preds)
```

```python
train_accuracy=accuracy_score(y_train,train_preds)
print(train_accuracy)
```

```python
confusion_matrix(y_test, test_preds)
```

```python
test_accuracy=accuracy_score(y_test,test_preds)
print(test_accuracy)
```

```python
from sklearn.metrics import recall_score
recall_score(y_train,train_preds)
```

```python
recall_score(y_test, test_preds)
```
